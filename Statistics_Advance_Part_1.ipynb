Q.1 What is a random variable in probability theory?
  - A random variable is a variable whose possible values are determined by chance events. It's a function that assigns a numerical value to each outcome in a sample space, allowing us to quantify and analyze uncertainty in a probabilistic framework. Random variables can be discrete (taking on specific, distinct values) or continuous (taking on any value within a given range or interval).

Q2. What are the types of random variables?
  - There are two main types of random variables:

    1. Discrete Random Variables: These can only take on a countable number of distinct values. Examples include:
        - The number of heads in a coin toss
        - The number of students in a class
        - The outcome of rolling a die

    2. Continuous Random Variables: These can take on any value within a given interval, or range. Examples include:
        - The height of a person
        - The time it takes to complete a task
        - The temperature on a given day

  These two types of random variables are used to model different kinds of uncertainty and are analyzed using different statistical methods.

Q3. What is the difference between discrete and continuous distributions?
  - Discrete Distributions:

  Data Type: Discrete data is countable and consists of separate, distinct values. Examples include the number of students in a class, the number of cars passing a point, or the outcomes of a dice roll.

Q4. What are probability distribution functions (PDF)?
  - A Probability Distribution Function (PDF) is a function that describes the probability of a random variable taking on specific values or falling within a particular range.

  For Discrete Random Variables:

    - A PDF, also known as a Probability Mass Function (PMF), assigns a probability to each possible value of the random variable.
    - The probabilities are non-negative and sum up to 1.

  For Continuous Random Variables:

    - A PDF describes the relative likelihood of the random variable taking on a particular value.
    - The probability of the random variable falling within a given interval is calculated by integrating the PDF over that interval.
    - The PDF is non-negative, and the area under the curve is equal to 1.

  In both cases, the PDF provides a way to calculate probabilities and understand the distribution of the random variable.

Q5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
  - Cumulative Distribution Functions (CDF) and Probability Density Functions (PDF) are two related but distinct concepts in probability theory.

  Key differences:

    - CDF (Cumulative Distribution Function):
        - Gives the probability that a random variable takes on a value less than or equal to a specific value.
        - Represents the cumulative probability up to a certain point.
        - Defined as F(x) = P(X ≤ x)
    - PDF (Probability Density Function):
        - Describes the probability distribution of a continuous random variable.
        - Gives the relative likelihood of a random variable taking on a specific value.
        - Defined as f(x) such that P(a ≤ X ≤ b) = ∫[a, b] f(x) dx

  Relationship between CDF and PDF:

    - The CDF is the integral of the PDF: F(x) = ∫(-∞, x] f(t) dt
    - The PDF is the derivative of the CDF: f(x) = dF(x)/dx

  In summary, the CDF provides the cumulative probability, while the PDF describes the probability distribution of a continuous random variable. Both are essential tools in probability theory and statistics.

Q7. What are the key properties of a Bernoulli distribution?
  - The Bernoulli distribution is a discrete probability distribution that models a single trial with two possible outcomes: success (1) or failure (0). The key properties of a Bernoulli distribution are:

  1. Two possible outcomes: The Bernoulli distribution has two possible outcomes, typically labeled as success (1) and failure (0).
  2. Probability of success: The probability of success (p) is a parameter of the distribution, where 0 ≤ p ≤ 1.
  3. Probability of failure: The probability of failure (q) is given by q = 1 - p.
  4. Expected value: The expected value of a Bernoulli distribution is p, which represents the probability of success.
  5. Variance: The variance of a Bernoulli distribution is p(1-p), which measures the spread of the distribution.

  The Bernoulli distribution is commonly used to model:

  - Binary outcomes: Yes/no, true/false, or success/failure outcomes.
  - Single trials: A single experiment or trial with two possible outcomes.

  The Bernoulli distribution is a fundamental building block for more complex distributions, such as the binomial distribution.

Q8. What is the binomial distribution, and how is it used in probability?

  The binomial distribution is used to calculate the probability of:

Q9. What is the Poisson distribution and where is it applied?

  The Poisson distribution of events is characterized by:

    - Average rate (λ): The average number of events occurring in the fixed interval.
    - Independence: Events occur independently of each other.

  The Poisson distribution is applied in various fields, including:

    - Modeling rare events: Such as accidents, natural disasters, or errors in manufacturing.
    - Queueing theory: Modeling the number of customers arriving at a service point.
    - Insurance: Modeling the number of claims or accidents.
    - Telecommunications: Modeling the number of phone calls or messages received.

  The Poisson distribution is useful when the number of events is large, but the probability of each event is small, making it a fundamental tool in statistics and data analysis.

Q10. What is a continuous uniform distribution?

    - Equal probability: Every value between a and b has the same probability density.

  f(x) = 1 / (b - a) for a ≤ x ≤ b

  The continuous uniform distribution is often used in simulations, modeling, and statistical analysis, particularly when there's no prior knowledge about the distribution of a variable within a certain range. Examples include:

    - Random number generation: Generating random numbers within a specific range.

Q11. What are the characteristics of a normal distribution?

    - 68-95-99.7 rule: About 68% of the data falls within 1 standard deviation of the mean, 95% within 2 standard deviations, and 99.7% within 3 standard deviations.

  The normal distribution is widely used in statistics, data analysis, and many fields, including:

    - Modeling real-world phenomena: Many natural phenomena, such as heights, weights, and IQ scores, follow a normal distribution.
    - Statistical inference: The normal distribution is used in hypothesis testing, confidence intervals, and regression analysis.

  The normal distribution is a fundamental concept in statistics, and its properties make it a powerful tool for data analysis and modeling.

Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
  - The Central Limit Theorem (CLT) states that:

    - Sample means converge to a normal distribution: As the sample size increases, the distribution of sample means approaches a normal distribution, regardless of the shape of the population distribution.

  The CLT is critical in statistics because:

    - Foundation for statistical inference: It provides a basis for using normal distribution-based methods for statistical inference, such as hypothesis testing and confidence intervals.
    - Large sample sizes: The CLT ensures that sample means are approximately normally distributed for large sample sizes, making it possible to use normal distribution-based methods.
    - Robustness: The CLT is robust to non-normality of the population distribution, allowing for the use of normal distribution-based methods even when the population distribution is not normal.

  The CLT has far-reaching implications in statistics, including:

    - Hypothesis testing: It enables the use of z-tests and t-tests for hypothesis testing.
    - Confidence intervals: It allows for the construction of confidence intervals for population means.
    - Regression analysis: It provides a foundation for regression analysis and modeling.

  The CLT is a fundamental concept in statistics, and its implications are essential for statistical analysis and inference.

    - Normal distribution as a limiting distribution: The normal distribution serves as a limiting distribution for the sample means, meaning that as the sample size becomes large, the distribution of sample means gets closer and closer to a normal distribution.

  In essence, the CLT explains why the normal distribution is so prevalent in statistics and why it can be used to approximate the distribution of sample means from any population distribution, given a large enough sample size.

  The CLT and normal distribution are closely linked, and this relationship is fundamental to many statistical methods, including:

Q15. What is the application of Z statistics in hypothesis testing?
  - The Z statistic is used in hypothesis testing to:

    - Compare a sample mean to a known population mean: When the population standard deviation is known and the sample size is large, the Z statistic can be used to test hypotheses about the population mean.
    - Calculate the probability of observing a sample mean: The Z statistic measures the number of standard deviations the sample mean is away from the population mean, allowing for the calculation of the probability of observing the sample mean (or a more extreme value) assuming the null hypothesis is true.

  The Z statistic is calculated as:

  Z = (x̄ - μ) / (σ / √n)

  Where:

    - x̄ is the sample mean
    - μ is the population mean
    - σ is the population standard deviation
    - n is the sample size

  The Z statistic is used in various hypothesis testing scenarios, including:

    - One-sample Z-test: Testing whether a sample mean is significantly different from a known population mean.
    - Two-sample Z-test: Testing whether the means of two independent samples are significantly different.

  The Z statistic is a powerful tool in hypothesis testing, allowing researchers to make informed decisions about population parameters based on sample data.

Q16. How do you calculate a Z-score, and what does it represent?
  - A Z-score, also known as a standard score, is calculated using the following formula:

  Z = (X - μ) / σ

  Where:

    - X is the value of the element
    - μ is the mean of the dataset
    - σ is the standard deviation of the dataset

  The Z-score represents:

    - Number of standard deviations away from the mean: A Z-score tells you how many standard deviations away from the mean a value is.
    - Relative position: A Z-score indicates the relative position of a value within a distribution.

  Interpretation of Z-scores:

    - Z = 0: The value is equal to the mean.
    - Z > 0: The value is above the mean.
    - Z < 0: The value is below the mean.
    - Z = 1: The value is 1 standard deviation above the mean.
    - Z = -1: The value is 1 standard deviation below the mean.

  Z-scores are useful for:

    - Comparing values: Z-scores allow for comparison of values from different distributions.
    - Identifying outliers: Z-scores can help identify values that are significantly different from the mean.
    - Standardizing data: Z-scores can be used to standardize data, making it easier to analyze and compare.

  In essence, Z-scores provide a way to understand the relative position of a value within a distribution and make comparisons across different distributions.

Q17. What are point estimates and interval estimates in statistics?
  - In statistics, estimates can be categorized into two main types:

    1. Point Estimates:
        - A point estimate is a single value used to estimate a population parameter.
        - It provides a specific value as the best estimate of the parameter.
        - Examples include the sample mean (x̄) as an estimate of the population mean (μ), or the sample proportion (p̂) as an estimate of the population proportion (p).

    2. Interval Estimates (Confidence Intervals):
        - An interval estimate provides a range of values within which the population parameter is likely to lie.
        - It gives a measure of the uncertainty or precision of the estimate.
        - Confidence intervals are constructed based on the point estimate and the variability of the sample data, often expressed with a certain level of confidence (e.g., 95% confidence interval).

  Key differences:
    - Point estimate: Provides a single value as an estimate.
    - Interval estimate: Provides a range of values, giving a sense of the estimate's precision.

  Both point and interval estimates are essential in statistics for making inferences about population parameters based on sample data.

Q19. What is the relationship between a Z-score and a confidence interval?
  - The relationship between a Z-score and a confidence interval is as follows:

    - Z-scores are used to construct confidence intervals: When constructing a confidence interval for a population mean, Z-scores are used to determine the margin of error.
    - Z-scores correspond to confidence levels: Specific Z-scores correspond to specific confidence levels. For example:
        - 95% confidence interval: Z = 1.96 (approximately)
        - 99% confidence interval: Z = 2.58 (approximately)

  The Z-score is used to calculate the margin of error, which is then added to and subtracted from the sample mean to construct the confidence interval.

  Formula: CI = x̄ ± (Z × σ / √n)

  Where:

    - x̄ is the sample mean
    - Z is the Z-score corresponding to the desired confidence level
    - σ is the population standard deviation
    - n is the sample size

  The relationship between Z-scores and confidence intervals allows researchers to quantify the uncertainty associated with their estimates and make informed decisions.

Q21. What are the assumptions for applying the Central Limit Theorem?
  - The Central Limit Theorem (CLT) relies on the following assumptions:

    1. Independence: The random variables being sampled must be independent of each other.
    2. Identical Distribution: The random variables being sampled should be identically distributed, meaning they have the same distribution, mean, and variance.
    3. Finite Variance: The random variables being sampled should have a finite variance.
    4. Sufficiently Large Sample Size: The sample size should be sufficiently large (typically n ≥ 30).

  If these assumptions are met, the CLT can be applied to approximate the distribution of the sample mean as a normal distribution, regardless of the shape of the population distribution.

  It's worth noting that some versions of the CLT can accommodate certain types of dependence or non-identical distributions, but these assumptions provide a general guideline for applying the CLT.

Q22. What is the concept of expected value in a probability distribution?
  - The expected value, also known as the mean or expectation, is a measure of the central tendency of a probability distribution. It represents the long-term average value that a random variable would take on if the experiment were repeated many times.

  The expected value is calculated as:

  E(X) = ∑xP(x) (for discrete random variables)
  E(X) = ∫xf(x)dx (for continuous random variables)

  Where:

    - E(X) is the expected value
    - x is the value of the random variable
    - P(x) is the probability of each value (for discrete random variables)
    - f(x) is the probability density function (for continuous random variables)

  The expected value has several important properties:

    - Long-term average: The expected value represents the long-term average value of a random variable.
    - Center of the distribution: The expected value is a measure of the center of the distribution.
    - Weighted average: The expected value is a weighted average of the possible values, where the weights are the probabilities.

  The expected value is a fundamental concept in probability theory and is used in a wide range of applications, including:

    - Decision-making under uncertainty: Expected values can be used to make informed decisions under uncertainty.
    - Risk analysis: Expected values can be used to quantify and manage risk.
    - Finance: Expected values are used to calculate expected returns on investments.

  In essence, the expected value provides a way to summarize the central tendency of a probability distribution and make informed decisions based on uncertain outcomes.

```python
# Q.2 Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

import numpy as np
import matplotlib.pyplot as plt

# Define the parameters of the discrete uniform distribution
a = 1  # Lower bound
b = 6  # Upper bound

# Generate the possible values of the random variable
x = np.arange(a, b + 1)

# Calculate the probability of each value
p = 1 / (b - a + 1)  # Probability of each value
y = np.repeat(p, b - a + 1)

# Plot the PMF
plt.figure(figsize=(8, 6))
plt.stem(x, y)
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Discrete Uniform Distribution PMF')
plt.xticks(x)
plt.ylim([0, p * 1.1])  # Set y-axis limit
plt.show()
```

```python
# Q.3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

import numpy as np

def bernoulli_pmf(x, p):
    """
    Calculate the probability mass function (PMF) of a Bernoulli distribution.

    Parameters:
    x (int): The value of the random variable (0 or 1).
    p (float): The probability of success (0 <= p <= 1).

    Returns:
    float: The probability of x.
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1")
    if p < 0 or p > 1:
        raise ValueError("p must be between 0 and 1")

    return p**x * (1-p)**(1-x)
# Example usage:
p = 0.7  # Probability of success
x_values = [0, 1]

for x in x_values:
    probability = bernoulli_pmf(x, p)
    print(f"P(X = {x}) = {probability}")
```

```python
# Q.4 Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.

import numpy as np
import matplotlib.pyplot as plt

# Set the parameters of the binomial distribution
n = 10  # Number of trials
p = 0.5  # Probability of success
num_simulations = 1000  # Number of simulations

# Simulate the binomial distribution
simulations = np.random.binomial(n, p, num_simulations)

# Calculate the mean and standard deviation
mean = np.mean(simulations)
std_dev = np.std(simulations)
print(f"Mean: {mean}, Standard Deviation: {std_dev}")

# Plot the histogram
plt.figure(figsize=(8, 6))
plt.hist(simulations, bins=np.arange(n + 2) - 0.5, align='mid', rwidth=0.8)
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xticks(np.arange(n + 1))
plt.show()
```

```python
# Q.5 Create a Paisson distribution and visualize it using Python.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Set the parameter of the Poisson distribution
lambda_ = 5  # Average rate of events

# Generate the possible values of the random variable
x = np.arange(0, 15)

# Calculate the probability of each value
probabilities = poisson.pmf(x, lambda_)

# Plot the PMF
plt.figure(figsize=(8, 6))
plt.plot(x, probabilities, 'o-', label=f'Poisson Distribution (λ={lambda_})')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.title('Poisson Distribution PMF')
plt.legend()
plt.show()

# Simulate the Poisson distribution
num_simulations = 1000
simulations = np.random.poisson(lambda_, num_simulations)

# Plot the histogram
plt.figure(figsize=(8, 6))
plt.hist(simulations, bins=np.arange(15) - 0.5, align='mid', rwidth=0.8, alpha=0.7, label='Simulated Data')
plt.plot(x, probabilities * num_simulations, 'ro-', label=f'Poisson Distribution (λ={lambda_})')
plt.xlabel('Number of Events')
plt.ylabel('Frequency')
plt.title('Poisson Distribution Histogram')
plt.legend()
plt.show()
```

```python
# Q.6 Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.

import numpy as np
import matplotlib.pyplot as plt

# Define the parameters of the discrete uniform distribution
a = 1  # Lower bound
b = 6  # Upper bound

# Generate the possible values of the random variable
x = np.arange(a, b + 1)

# Calculate the probability of each value
p = 1 / (b - a + 1)  # Probability of each value

# Calculate the CDF
cdf = np.cumsum(np.repeat(p, b - a + 1))

# Plot the CDF
plt.figure(figsize=(8, 6))
plt.step(x, cdf, where='post')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('Discrete Uniform Distribution CDF')
plt.xticks(x)
plt.ylim([0, 1.1])  # Set y-axis limit
plt.grid(True)
plt.show()
```

```python
# Q.7 Generate a continuous uniform distribution using NumPy and visualize it.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Define the parameters of the continuous uniform distribution
a = 0  # Lower bound
b = 1  # Upper bound

# Generate the possible values of the random variable
x = np.linspace(a - 1, b + 1, 400)

# Calculate the probability density function (PDF)
pdf = uniform.pdf(x, loc=a, scale=b-a)

# Plot the PDF
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f'Uniform Distribution (a={a}, b={b})')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Continuous Uniform Distribution PDF')
plt.legend()
plt.show()
```

```python
# Q.9 Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    # Calculate the mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate the Z-scores
    z_scores = (data - mean) / std_dev

    return z_scores

def plot_z_scores(data, z_scores):
    # Plot the original data
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.hist(data, bins=30, alpha=0.7)
    plt.title('Original Data')
    plt.xlabel('Value')
    plt.ylabel('Frequency')

    # Plot the Z-scores
    plt.subplot(1, 2, 2)
    plt.hist(z_scores, bins=30, alpha=0.7)
    plt.title('Z-scores')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.axvline(x=-2, color='red', linestyle='--', label='-2 standard deviations')
    plt.axvline(x=2, color='red', linestyle='--', label='2 standard deviations')
    plt.legend()
    plt.show()

# Generate a random dataset
np.random.seed(0)
data = np.random.normal(0, 1, 1000)

# Calculate and plot the Z-scores
z_scores = calculate_z_scores(data)
plot_z_scores(data, z_scores)
```

```python
# Q.10 Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Define the parameters of the exponential distribution
scale = 1  # Scale parameter (1/lambda)

# Generate random samples from the exponential distribution
num_samples = 1000
sample_size = 30  # Sample size for the CLT
num_trials = 1000  # Number of trials

# Generate samples and calculate the sample means
sample_means = np.array([np.mean(expon.rvs(scale=scale, size=sample_size)) for _ in range(num_trials)])

# Calculate the mean and standard deviation of the sample means
mean_sample_means = np.mean(sample_means)
std_dev_sample_means = np.std(sample_means)
print(f"Mean of Sample Means: {mean_sample_means}, Standard Deviation of Sample Means: {std_dev_sample_means}")

# Plot the histogram of the sample means
plt.figure(figsize=(8, 6))
plt.hist(sample_means, bins=30, density=True, alpha=0.7, label='Sample Means')

# Plot the normal distribution PDF
x = np.linspace(mean_sample_means - 3 * std_dev_sample_means, mean_sample_means + 3 * std_dev_sample_means, 100)
pdf = np.exp(-((x - mean_sample_means) / std_dev_sample_means)**2 / 2) / (std_dev_sample_means * np.sqrt(2 * np.pi))
plt.plot(x, pdf, 'r-', label='Normal Distribution')

plt.xlabel('Sample Mean')
plt.ylabel('Probability Density')
plt.title('Central Limit Theorem Demonstration')
plt.legend()
plt.show()
```

```python
# Q.12 Write a Python function to calculate and plot the standard normal distrinution (mean = 0, std = 1).

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal_distribution():
    # Define the range of x values
    x = np.linspace(-3, 3, 100)

    # Calculate the probability density function (PDF)
    pdf = norm.pdf(x, loc=0, scale=1)

    # Calculate the cumulative distribution function (CDF)
    cdf = norm.cdf(x, loc=0, scale=1)

    # Create a figure with subplots
    fig, axs = plt.subplots(2, figsize=(8, 8))

    # Plot the PDF
    axs[0].plot(x, pdf)
    axs[0].set_title('Probability Density Function (PDF)')
    axs[0].set_xlabel('x')
    axs[0].set_ylabel('Probability Density')

    # Plot the CDF
    axs[1].plot(x, cdf)
    axs[1].set_title('Cumulative Distribution Function (CDF)')
    axs[1].set_xlabel('x')
    axs[1].set_ylabel('Cumulative Probability')

    plt.tight_layout()
    plt.show()

plot_standard_normal_distribution()
```

```python
# Q.14 Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

def calculate_z_score(data_point, mean, std_dev):
    # Calculate the Z-score
    z_score = (data_point - mean) / std_dev
    return z_score

def plot_standard_normal_distribution(z_score):
    # Define the range of x values
    x = np.linspace(-3, 3, 100)

    # Calculate the probability density function (PDF)
    pdf = norm.pdf(x, loc=0, scale=1)

    # Plot the PDF
    plt.figure(figsize=(8, 6))
    plt.plot(x, pdf)
    plt.axvline(x=z_score, color='red', linestyle='--', label=f'Z-score = {z_score:.2f}')
    plt.fill_between(x, pdf, where=(x >= z_score), color='blue', alpha=0.3, label='Right tail')
    plt.fill_between(x, pdf, where=(x <= z_score), color='green', alpha=0.3, label='Left tail')
    plt.legend()
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.title('Standard Normal Distribution')
    plt.show()

    # Calculate the probability
    probability = norm.cdf(z_score)
    print(f'P(Z <= {z_score:.2f}) = {probability:.4f}')
    print(f'P(Z > {z_score:.2f}) = {1 - probability:.4f}')

# Define the data point, mean, and standard deviation
data_point = 85
mean = 80
std_dev = 5

# Calculate the Z-score
z_score = calculate_z_score(data_point, mean, std_dev)
print(f'Z-score = {z_score:.2f}')

# Plot the standard normal distribution
plot_standard_normal_distribution(z_score)
```

```python
# Q.15 Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

# Define the null and alternative hypotheses
null_hypothesis = "The average height of adults in the population is 175 cm"
alternative_hypothesis = "The average height of adults in the population is not 175 cm"

# Define the sample dataset
sample_heights = np.array([170, 180, 165, 178, 182, 168, 175, 172, 179, 181])

# Calculate the sample mean and standard deviation
sample_mean = np.mean(sample_heights)
sample_std_dev = np.std(sample_heights)

# Define the population mean and standard deviation
population_mean = 175
population_std_dev = 5  # Known population standard deviation

# Calculate the Z-statistic
n = len(sample_heights)
z_statistic = (sample_mean - population_mean) / (population_std_dev / np.sqrt(n))

# Calculate the p-value
p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

# Define the significance level
alpha = 0.05

# Print the results
print(f"Null Hypothesis: {null_hypothesis}")
print(f"Alternative Hypothesis: {alternative_hypothesis}")
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Make a decision
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
# Q.16 Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy import stats

# Define the dataset
data = np.array([23, 21, 19, 24, 20, 22, 18, 25, 19, 21, 20, 22])

# Calculate the sample mean and standard deviation
sample_mean = np.mean(data)
sample_std_dev = np.std(data, ddof=1)  # Use sample standard deviation

# Define the confidence level
confidence_level = 0.95

# Calculate the standard error
n = len(data)
standard_error = sample_std_dev / np.sqrt(n)

# Calculate the critical value
critical_value = stats.t.ppf((1 + confidence_level) / 2, n - 1)

# Calculate the margin of error
margin_of_error = critical_value * standard_error

# Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Print the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Interval ({confidence_level*100}%): ({lower_bound:.2f}, {upper_bound:.2f})")

# Interpret the result
print(f"We are {confidence_level*100}% confident that the true population mean lies between {lower_bound:.2f} and {upper_bound:.2f}.")
```

```python
# Q.18 Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the parameters of the normal distribution
mu = 0  # Mean
sigma = 1  # Standard deviation

# Generate x values
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)

# Calculate the probability density function (PDF)
pdf = norm.pdf(x, loc=mu, scale=sigma)

# Plot the PDF
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f'Normal Distribution (μ={mu}, σ={sigma})')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution PDF')
plt.legend()
plt.fill_between(x, pdf, alpha=0.2)
plt.show()

# Calculate and print the probability for a specific range
lower_bound = -1
upper_bound = 1
probability = norm.cdf(upper_bound, loc=mu, scale=sigma) - norm.cdf(lower_bound, loc=mu, scale=sigma)
print(f'P({lower_bound} <= X <= {upper_bound}) = {probability:.4f}')

# Plot the PDF with different standard deviations
plt.figure(figsize=(8, 6))
for sigma in [0.5, 1, 2]:
    pdf = norm.pdf(x, loc=mu, scale=sigma)
    plt.plot(x, pdf, label=f'σ={sigma}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution PDF with Different Standard Deviations')
plt.legend()
plt.show()
```

```python
# Q.19 Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt

# Define the parameter of the Poisson distribution
lambda_param = 5  # Average rate of events

# Generate x values
x = np.arange(0, 15)

# Calculate the probability mass function (PMF)
pmf = poisson.pmf(x, mu=lambda_param)

# Calculate the cumulative distribution function (CDF)
cdf = poisson.cdf(x, mu=lambda_param)

# Print the CDF values
print("x\tCDF")
for i in range(len(x)):
    print(f"{x[i]}\t{cdf[i]:.4f}")

# Plot the CDF
plt.figure(figsize=(8, 6))
plt.plot(x, cdf, marker='o')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.title(f'Poisson Distribution CDF (λ={lambda_param})')
plt.show()

# Calculate and interpret the probability for a specific range
lower_bound = 3
upper_bound = 7
probability = poisson.cdf(upper_bound, mu=lambda_param) - poisson.cdf(lower_bound - 1, mu=lambda_param)
print(f'P({lower_bound} <= X <= {upper_bound}) = {probability:.4f}')

# Calculate and interpret the probability for a specific value
x_value = 5
probability = poisson.pmf(x_value, mu=lambda_param)
print(f'P(X = {x_value}) = {probability:.4f}')

# Calculate and interpret the probability that X is greater than a specific value
x_value = 10
probability = 1 - poisson.cdf(x_value, mu=lambda_param)
print(f'P(X > {x_value}) = {probability:.4f}')
```

```python
# Q.22 Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

import numpy as np

# Generate a dataset from a normal distribution
np.random.seed(0)
mu = 0  # Mean
sigma = 1  # Standard deviation
n = 1000  # Number of data points
dataset = np.random.normal(mu, sigma, n)

# Calculate the range
data_min = np.min(dataset)
data_max = np.max(dataset)
range_value = data_max - data_min

# Calculate the interquartile range (IQR)
q1 = np.percentile(dataset, 25)
q3 = np.percentile(dataset, 75)
iqr = q3 - q1

# Print the results
print(f"Range: {range_value:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
print(f"Q1 (25th percentile): {q1:.2f}")
print(f"Q3 (75th percentile): {q3:.2f}")

# Interpret the results
print("The range represents the difference between the largest and smallest values in the dataset.")
print("The IQR represents the difference between the 75th percentile (Q3) and the 25th percentile (Q1) of the dataset.")
print("The IQR is a measure of the spread of the middle 50% of the data.")
```

```python
# Q.24 Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

# Generate a dataset from a normal distribution
np.random.seed(0)
mu = 0  # Mean
sigma = 1  # Standard deviation
n = 1000  # Number of data points
dataset = np.random.normal(mu, sigma, n)

# Calculate the skewness
skewness_value = skew(dataset)

# Calculate the kurtosis (excess kurtosis)
kurtosis_value = kurtosis(dataset)

# Calculate the kurtosis (actual kurtosis)
actual_kurtosis = kurtosis(dataset) + 3

# Print the results
print(f"Skewness: {skewness_value:.4f}")
print(f"Excess Kurtosis: {kurtosis_value:.4f}")
print(f"Actual Kurtosis: {actual_kurtosis:.4f}")

# Interpret the results
if skewness_value > 0:
    print("The distribution is positively skewed.")
elif skewness_value < 0:
    print("The distribution is negatively skewed.")
else:
    print("The distribution is symmetric.")

if kurtosis_value > 0:
    print("The distribution is leptokurtic (heavy-tailed).")
elif kurtosis_value < 0:
    print("The distribution is platykurtic (light-tailed).")
else:
    print("The distribution is mesokurtic (normal-tailed).")
```
